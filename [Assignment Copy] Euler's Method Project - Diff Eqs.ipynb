## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
p = -1
q = 3

def Fx(v):
    return v

def Fv(x, v):
    return -p*v - q*x
```

## Initial Values

```python
x0 = -4       # initial condition for x(t)
v0 = 4        # initial condition for v(t)

dt = .0001    # time increment
T = 5         # total time over which to approximate solution
```

## Euler's Method Calculations (Cont. Updates)

```python
FtList = np.linspace(0, T, int(T/dt) + 1)
FxList = np.zeros(len(FtList))
FvList = np.zeros(len(FxList))

FxList[0] = x0
FvList[0] = v0

for i in range(1, len(FxList)):
    
    FxList[i], FvList[i] = FxList[i-1] + Fx(FvList[i-1])*dt, FvList[i-1] + Fv(FxList[i-1], FvList[i-1])*dt
```

```python
for i in range(len(FxList)):
    print(i+1, '= (', round(FxList[i], 3), ' ,', round(FvList[i], 3),')')
```

## Actual Solution Values

```python
print(FxList[-1], FvList[-1]) # "highest" of the 50,001 values in FxList and FvList, respectively
```

## Plotting

```python
plt.subplot(111)

plt.plot(FxList, FvList, 'c-',  label="Euler's Method")

plt.axis([-80,80,-100,100])

plt.xlabel('x(t)')
plt.ylabel('v(t)')
plt.legend(bbox_to_anchor=(0, 1.02, 1, .102), loc=3, ncol=2, 
           mode="expand", borderaxespad=0)

plt.show()
```
